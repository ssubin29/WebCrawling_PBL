```python
#클래스

class Qurd():
    width=0
    height=0
    color=''
    name='Qurd'
    # 메서드 인자는 모두 사용 안해도 된다
    def qurd_name(self, argument1):
        print(argument1)
        return self.name
```

```python
quad1= Qurd()
quad2=Qurd()
```

```python
quad1.width=10
quad1.height=10
quad1.color='blue'
quad1.name='blue 사각형'
```

```python
print (quad1.width,quad2.width)
```

```python
quad2.width=5
quad2.height=5
quad2.color='red'
quad2.name='red 사각형'
```

```python
print(quad2.qurd_name(1))
```

```python
#크롤링 패턴 코드 

import requests 
from bs4 import BeautifulSoup

res = requests.get("https://www.inflearn.com/course/python-crawling-basic/lecture/21266?tab=note&speeed")
soup=BeautifulSoup(res.content,'html.parser')
mydata=soup.find('title')
print(mydata.get_text())
```

```python
html="""<html>
            <body>
            <h1 id='title'> [1]크롤링이란?</h1>
            <p class ='cssstyle'> 웹 페이지에서 필요한 데이터를 추출하는 것</p>
            <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
            
            </body>
</html>

"""

soup=BeautifulSoup(html,'html.parser')

#태그로 검색 방법
data=soup.find('h1')
print(data)
print(data.string)
```

```python
data2=soup.find('p')
print(data2)
print(data2.string)
```

```python
data3=soup.find(id='body')
print(data3)
print(data3.string)
```

```python
data4=soup.find_all('p')
print(data4)
for item in data4:
    print(item.string)
```
