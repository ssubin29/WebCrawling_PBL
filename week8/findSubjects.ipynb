```python
#현재 학점 관리 클래스 


class Credit:
    
    def __init__(self,major):
        self.major=major
        self.core_major=0
        self.advanced_major=0
        self.ilban_gyoyang=0
        self.haegsim_gyoyang=0
        self.gongtong_gyoyang=0
    
    def addMajor(self,major_type,major):
        if major_type=='복수전공':
            self.minor=-1
            self.double_major_core=0
            self.double_major_advanced=0
        if major_type=='부전공':
            self.minor=0
            self.double_major_core=-1
            self.double_major_advanced=-1
    
    def addCredit(self,major,subject_type,point):
        
        if subject_type=='핵심전공':
            if self.major==major:
                self.core_major+=int(point)
            elif self.minor==-1:
                self.double_major_corer+=int(point)
            else:
                self.minor+=int(point)
                
        elif subject_type=='심화전공':
            if self.major==major:
                self.advanced_major+=int(point)
            elif self.minor==-1:
                self.double_major_advanced+=int(point)
            else:
                self.minor+=int(point)
                
        elif subject_type=='일반교양':
            self.ilban_gyoyang+=int(point)
        elif subject_type=='핵심교양':
            self.haegsim_gyoyang+=int(point)
        elif subject_type=='공통교양':
            self.gongtong_gyoyang+=int(point)
            
        else:
            print('type 오류')
    
    
    def show(self):
        print("-------- 전공 ---------")
        print("핵심전공 : "+str(self.core_major))
        print("심화전공 : "+str(self.advanced_major))
        print("-------- 교양 ---------")
        print("일반교양 : "+str(self.ilban_gyoyang))
        print("핵심교양 : "+str(self.haegsim_gyoyang))
        print("공통교양: "+str(self.gongtong_gyoyang))
        if self.minor==-1:
            print("-------- 복수전공 ---------")
            print("핵심전공 : "+str(self.double.major.core))
            print("심화전공 : "+str(self.double.major.advanced))
        else:
            print("-------- 부전공 ---------")
            print("부전공 : "+str(self.minor))
                
    
```
