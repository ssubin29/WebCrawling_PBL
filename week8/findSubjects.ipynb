```python
#현재 학점 관리 클래스 


class Credit:
    
    def __init__(self,id,major):
        self.id=int(id/10000%100)
        self.major=major
        self.core_major=0
        self.advanced_major=0
        self.ilban_gyoyang=0
        self.haegsim_gyoyang=0
        self.gongtong_gyoyang=0
        self.total=0
        self.teaching_profession=-1
        
    
    def addMajor(self,major_type,major):
        if major_type=='복수전공':
            self.minor=-1
            self.double_major_core=0
            self.double_major_advanced=0
        if major_type=='부전공':
            self.minor=0
            self.double_major_core=-1
            self.double_major_advanced=-1
        if major_type=='교직이수':
            self.teaching_profession=0
    
    def addCredit(self,major,subject_type,point):
        
        if subject_type=='핵심전공':
            if self.major==major:
                self.core_major+=int(point)
                self.total+=int(point)
            elif self.minor==-1:
                self.double_major_corer+=int(point)
                self.total+=int(point)
            else:
                self.minor+=int(point)
                self.total+=int(point)
                
        elif subject_type=='심화전공':
            if self.major==major:
                self.advanced_major+=int(point)
                self.total+=int(point)
            elif self.minor==-1:
                self.double_major_advanced+=int(point)
                self.total+=int(point)
            else:
                self.minor+=int(point)
                self.total+=int(point)
                
        elif subject_type=='일반교양' or(subject_type=='진로소양' and (self.id>=20)):
            self.ilban_gyoyang+=int(point)
            self.total+=int(point)
        elif subject_type=='핵심교양' or (subject_type=='진로소양' and (self.id<20)):
            self.haegsim_gyoyang+=int(point)
            self.total+=int(point)
        elif subject_type=='공통교양':
            self.gongtong_gyoyang+=int(point)
            self.total+=int(point)
        elif subject_type=='교직':
            self.teaching_profession+=int(point)
            self.total+=int(point)
            
        else:
            print('type 오류')
    
    
    def show(self):
        print("-------- 전공 ---------")
        print("핵심전공 : "+str(self.core_major))
        print("심화전공 : "+str(self.advanced_major))
        print("-------- 교양 ---------")
        print("일반교양 : "+str(self.ilban_gyoyang))
        print("핵심교양 : "+str(self.haegsim_gyoyang))
        print("공통교양: "+str(self.gongtong_gyoyang))
        if self.minor==-1:
            print("-------- 복수전공 ---------")
            print("핵심전공 : "+str(self.double.major.core))
            print("심화전공 : "+str(self.double.major.advanced))
        else:
            print("-------- 부전공 ---------")
            print("부전공 : "+str(self.minor))
        if self.teaching_profession>=0:
            print("-------- 교직 ---------")
            print("교직 이수: "+str(self.teaching_profession))
            
    def calculation(self,college,major,gg,hg,ig,total_g,cm,am,total_m,s_cm,s_am,total_s,minor,d_cm,d_am,total_d,teaching,total):
        print (college + " / "+ major )
        print("-------- 전공 ---------")
        print("핵심전공 : "+str(self.core_major)+' / '+cm) 
        print("심화전공 : "+str(self.advanced_major)+' / '+am)
        print("계 : ")
        print("-------- 교양 ---------")
        print("일반교양 / 진로소양 : "+str(self.ilban_gyoyang)+' / '+ig)
        print("핵심교양 : "+str(self.haegsim_gyoyang)+' / '+hg)
        print("공통교양: "+str(self.gongtong_gyoyang)+' / '+gg)
        
        if self.minor==-1:
            print("-------- 복수전공 ---------")
            print("핵심전공 : "+str(self.double.major.core)+' / '+d_cm)
            print("심화전공 : "+str(self.double.major.advanced)+' / '+d_am)
        else:
            print("-------- 부전공 ---------")
            print("부전공 : "+str(self.minor)+' / '+minor)
            
        if self.teaching_profession>=0:
            print("-------- 교직 ---------")
            print("교직 이수: "+str(self.teaching_profession)+' / '+teaching)
        print('---------------------------------------')  
        print("총 학점 : "+str(self.total))
            
        print("========================================================================")
        print("졸업을 위해 필요한 학점")
        if cm=='0' and am=='0':
            print('전공 : '+str(int(total_m)-self.core_major-self.advanced_major))
        else:
            print("-------- 전공 ---------")
            print("핵심전공 : "+str(int(cm)-self.core_major if (int(cm)-self.core_major) >0 else "충족"  )) 
            print("심화전공 : "+str(int(am)-self.advanced_major if (int(am)-self.advanced_major) >0 else "충족"))
        
        if (self.core_major >= (int(cm) +int(s_cm))) and (self.advanced_major >=(int(am)+int(s_am))):
            print("전공 심화가 인정되었습니다.")
        
        print("-------- 교양 ---------")
        print("일반교양 /진로 소양: "+str(int(ig)-self.ilban_gyoyang if (int(ig)-self.ilban_gyoyang) >0 else "충족"))
        print("핵심교양 : "+str(int(hg)-self.haegsim_gyoyang if (int(hg)-self.haegsim_gyoyang) >0 else "충족"))
        print("공통교양: "+str(int(gg)-self.gongtong_gyoyang if (int(gg)-self.gongtong_gyoyang) >0 else "충족"))
        if self.minor==-1:
            print("-------- 복수전공 ---------")
            print("핵심전공 : "+str(int(d_cm)-self.double.major.core if (int(d_cm)-self.double.major.core) >0 else "충족"))
            print("심화전공 : "+str(int(d_am)-self.double.major.advanced if (int(d_am)-self.double.major.advanced) >0 else "충족"))
        else:
            print("-------- 부전공 ---------")
            print("부전공 : "+str(int(minor)-self.minor if (int(minor)-self.minor) >0 else "충족"))
            
        if self.teaching_profession>=0:
            print("-------- 교직 ---------")
            print("교직 이수: "+str(int(teaching)-self.teaching_profession if (int(teaching)-self.teaching_profession) >0 else "충족"))
        print('---------------------------------------')    
        
                        
            
        
            
        
    
```
