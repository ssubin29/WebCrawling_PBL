```python
import requests
from bs4 import BeautifulSoup

exchange_rate={}
for page_num in range(10):
    res = requests.get('https://finance.naver.com/marketindex/exchangeDailyQuote.nhn?marketindexCd=FX_USDKRW&page=' + str(page_num + 1))
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('table.tbl_exchange.today tbody tr')
    for item in data:
        exchange_rate[item.find('td',{'class':'date'}).get_text().strip()] = float(item.find('td',{'class':'num'}).get_text().strip().replace(',',''))   
#print(exchange_rate)
purchase_amount=1113
purchase_date={}
for date in exchange_rate:
    if exchange_rate[date]<1113 :
        purchase_date[date]=exchange_rate[date]
        
print(purchase_date)
```

```python
import sys
from PyQt5.QtWidgets import QApplication, QWidget, QLabel, QVBoxLayout, QCalendarWidget
from PyQt5.QtCore import QDate
import requests
from bs4 import BeautifulSoup


class MyApp(QWidget):

    def __init__(self,date_dic):
        super().__init__()
        self.date_dic
        self.initUI()

    def initUI(self):
        cal = QCalendarWidget(self)
        cal.setGridVisible(True)
        cal.clicked[QDate].connect(self.showDate)

        self.lbl = QLabel(self)
        date = cal.selectedDate()
        self.lbl.setText(date.toString())

        vbox = QVBoxLayout()
        vbox.addWidget(cal)
        vbox.addWidget(self.lbl)

        self.setLayout(vbox)

        self.setWindowTitle('QCalendarWidget')
        self.setGeometry(300, 300, 400, 300)
        self.show()
        
        

    def showDate(self, date):
        self.lbl.setText(date.toString())


if __name__ == '__main__':
   
    
    # 크롤링하는 부분 
    exchange_rate={} #전체 환율 dic
    for page_num in range(10):
        res = requests.get('https://finance.naver.com/marketindex/exchangeDailyQuote.nhn?marketindexCd=FX_USDKRW&page=' + str(page_num + 1))
        soup = BeautifulSoup(res.content, 'html.parser')
        data = soup.select('table.tbl_exchange.today tbody tr')
        for item in data:
            exchange_rate[item.find('td',{'class':'date'}).get_text().strip()] = float(item.find('td',{'class':'num'}).get_text().strip().replace(',',''))   
    #print(exchange_rate)
    purchase_amount=1113 # 목표 구매 가격 --> 이후 사용자 입력을 받고 싶음
    purchase_date={}     # 구매 가능 날짜 dic 
    
    for date in exchange_rate:
        if exchange_rate[date]<1113 :
            purchase_date[date]=exchange_rate[date]

    print(purchase_date)
    
    #Gui 사용 
    app = QApplication(sys.argv)
    ex = MyApp(purchase_date)
    sys.exit(app.exec_())
    
    
```
